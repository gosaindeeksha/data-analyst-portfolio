PROBLEM:Department Highest Salary

```python
import pandas as pd

def department_highest_salary(employee: pd.DataFrame, department: pd.DataFrame) -> pd.DataFrame:
    employee = employee.merge(
    department[['id', 'name']],
    left_on='departmentId',
    right_on='id',
    how='left')
    employee = employee.rename(columns={'name_y': 'Department','name_x':'Employee','salary':'Salary'})
    employee['MaxSalary']=employee.groupby('Department')['Salary'].transform('max')
    return employee[employee['MaxSalary']==employee['Salary']][['Department','Employee','Salary']]

    
```
