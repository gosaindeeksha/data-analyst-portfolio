PROBLEM:Nth Highest Salary

```python
import pandas as pd

def nth_highest_salary(employee: pd.DataFrame, N: int) -> pd.DataFrame:
    uni_salaries = employee['salary'].drop_duplicates().sort_values(ascending=False)
    if len(uni_salaries)>=N and N>0:
        return pd.DataFrame({f'getNthHighestSalary({N})':[uni_salaries.iloc[N-1]]})
    else:
        return pd.DataFrame({f'getNthHighestSalary({N})':[None]})
```
