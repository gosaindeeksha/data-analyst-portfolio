PROBLEM:Calculate Special Bonus

```python
import pandas as pd

def calculate_special_bonus(employees: pd.DataFrame) -> pd.DataFrame:
    employees['bonus'] = 0
    is_true = (employees['employee_id']%2 ==1) & (employees['name'].str[0] != 'M')
    employees.loc[is_true,'bonus'] = employees.loc[is_true,'salary']
    return employees[['employee_id','bonus']].sort_values('employee_id')
```
