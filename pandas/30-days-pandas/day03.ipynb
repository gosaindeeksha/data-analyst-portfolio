PROBLEM:Customers Who Never Order

```python
import pandas as pd

def find_customers(customers: pd.DataFrame, orders: pd.DataFrame) -> pd.DataFrame:
    nevcustom = list(set(customers['id']).difference(set(orders['customerId'])))
    return customers[customers['id'].isin(nevcustom)][['name']].rename(columns={'name':'Customers'})
```
