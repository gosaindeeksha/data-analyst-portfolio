PROBLEM:Second Highest Salary

```python
import pandas as pd

def second_highest_salary(employee: pd.DataFrame) -> pd.DataFrame:
    uni_salaries=employee['salary'].drop_duplicates().sort_values(ascending=False)
    if len(uni_salaries)>=2:
        return pd.DataFrame({'SecondHighestSalary':[uni_salaries.iloc[1]]})
    else:
        return pd.DataFrame({'SecondHighestSalary':[None]})
```
